```python
##Please find top 5 films in movie_performance dataframe with highest "Gross to Budget" value (earning ratio) and create a bar plot that shows film names on the x-axis and “Gross to Budget” on the y-axi
import matplotlib.pyplot as plt
import pandas as pd
df = pd.read_csv('C:\\Users\\kasie\\OneDrive\\Pulpit\\MCU2.csv', encoding = 'unicode_escape')

top5 = df[['Film', 'Gross to Budget']].sort_values(by = 'Gross to Budget', ascending = False).head(5)

fig,ax = plt.subplots(figsize = (12,6)) 
films = top5['Film']
earning_ratio = top5['Gross to Budget']

bar_labels = ['red', 'blue', 'green', 'orange', 'purple']
bar_colors = ['tab:red', 'tab:blue', 'tab:green', 'tab:orange', 'tab:purple']

ax.bar (top5['Film'], top5['Gross to Budget'], label=bar_labels, color=bar_colors)
ax.set_xlabel ('Flms')
ax.set_ylabel('Gross to Budget')
ax.set_title ('Top five movies by earning ratio')
ax.legend (title = 'Film color')
plt.show


## Please create a pie chart, displaying the portion of films falling into each of “Recommended” column categories
percentages = df['Recommend'].value_counts(normalize = True).rename('Percentages').reset_index() 
#matplotlib
sizes = percentages ['Percentages']
labels = percentages ['Recommend']

fix,ax = plt.subplots()
ax.pie(sizes, labels=labels, autopct='%1.1f%%') 

print (percentages)

#pandas
df['Recommend'].value_counts().plot(kind='pie', autopct='%1.1f%%', colors = ['yellow', 'black', 'pink'], title='Percentage of recommended movies')
plt.ylabel('')
plt.show
```
