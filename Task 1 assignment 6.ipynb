```python
import pandas as pd

## Create a dataframe, called movie_performance, with dc_marvel_movie_performance.csv data and check how data looks like (play with head(), tail(), check column names etc)
df = pd.read_csv ('C:\\Users\\kasie\\Downloads\\dc_marvel_movie_performance.csv', encoding = 'unicode_escape') 
df.tail()
print (df.tail(2))

## Remove MCU, Phase and Franchise columns
columnsToDelete = ['MCU', 'Phase', 'Franchise']
df.drop(columns = columnsToDelete, inplace = True) 


## Find unique values of Character Family column and print them sorted from A to Z
uniqueFamily = df['Character Family'].unique() 
uniqueFamily.sort() 
print (uniqueFamily) 

## Find longest movie
df['Length'] = df['Length']
longestMovie = df.loc[df['Length'].idxmax()]
print ('The longest and the coolest Marvel movie is: ', longestMovie['Film']) 

## Find mean values of “Rotten Tomatoes Critic Score” per “Break Even” categories 
BestTeam = df.groupby('Break Even')['Rotten Tomatoes Critic Score']
print ( 'Average score is: \n', round(BestTeam.mean(),1))

## Add a new column to movie_performance dataframe,  called “Recommended”, which takes value from warunki yes, no 
def recommendMovie(Rotten_Tomatoes_Critic_Score):
    if Rotten_Tomatoes_Critic_Score  >= 90:
        return 'Strongly Yes'
    elif 60 < Rotten_Tomatoes_Critic_Score < 90:
        return 'Yes'
    elif 40 < Rotten_Tomatoes_Critic_Score <= 59:
        return 'NO'
    elif Rotten_Tomatoes_Critic_Score < 40:
        return 'Strongly NO'
    else: 
        return None
df['Recommend'] = df['Rotten Tomatoes Critic Score'].apply(recommendMovie)
data = {'Recommend'} 
percentage = df['Recommend'].value_counts(normalize = False) 
print ("Percentages are: \n", percentage)
df.to_csv('C:\\Users\\kasie\\OneDrive\\Pulpit\\MCU2.csv' , index = False) 
```
